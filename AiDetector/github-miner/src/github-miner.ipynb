```python
%load_ext autoreload
%autoreload 2

import json

from github.Repository import RepositorySearchResult

from pipelines.shared import input_until_integer

from dotenv import load_dotenv
from github import Auth, Github
import os
from git import Repo

load_dotenv()
TOKEN_ENV = "GITHUB_TOKEN"

auth = Auth.Token(os.getenv(TOKEN_ENV))
gh = Github(auth=auth)
```

```python
# Query repos

query = input("Enter query or leave blank for default: ")

if len(query) == 0:
    query = "language:java license:mit license:apache-2.0 springboot"
page_number = input_until_integer("Enter page number: ")

search_result = gh.search_repositories(query=query)
page_results = search_result.get_page(page_number)


def map_to_dict(result: RepositorySearchResult):
    return {
        "clone_url": result.clone_url,
        "name": result.name,
        "license": result.license.name,
    }


dict_list = list(map(map_to_dict, page_results))
print(dict_list)

output_path = input("Please enter output file: ")

with open(output_path, "w") as file:
    file.write(json.dumps(dict_list, indent=2))
```

```python
# Clone repos

repos_file = input("Enter repos path: ")

with open(repos_file, "r") as file:
    repos = json.load(file)

local_url = input("Enter local clone url: ")


def clone_repo(repo):
    repo_name = repo["name"]
    repo_path = local_url + repo_name
    Repo.clone_from(repo["clone_url"], repo_path, depth=1)
    return repo_path


for repository in repos:
    clone_repo(repository)
```
